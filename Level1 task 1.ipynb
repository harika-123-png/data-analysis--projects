```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("dataset3.csv")
data.head
```

```python
data_cuisines=data["Cuisines"].dropna()
```

```python
cuisine_count=data["Cuisines"].str.split(',').explode().value_counts()
top_three_cuisine=cuisine_count.head(3)
print(top_three_cuisine)
```

```python
colors=['green','yellow','red']
plt.bar(top_three_cuisine.index,top_three_cuisine.values,color=colors)
plt.title("The Top Three Most Common Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Count")
plt.show()
```

```python
total_restaurants =len(data_cuisines)
top_cuisines_percentage=(top_three_cuisine/total_restaurants)*100
```

```python
plt.pie(top_three_cuisine.values,labels=top_three_cuisine.index,autopct='%1.1f%%')
plt.title("The Percentage of Restaurants That Serve Each of The Top Cuisines")
plt.show()
```
